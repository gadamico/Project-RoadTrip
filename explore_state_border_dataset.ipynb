```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('borders.db')
```

```python
pd.read_sql(
    """
    SELECT *
    FROM borders
    WHERE Interstate = 0
    ORDER BY LengthRank
    """
, conn
)
```

```python
pd.read_sql(
    """
    SELECT *
    FROM borders
    WHERE TimeZone = 'Pacific'
    ORDER BY LengthRank
    """
, conn
)
```

```python
pd.read_sql(
    """
    SELECT *
    FROM borders
    WHERE River = 'Mississippi'
    ORDER BY LengthRank
    """
, conn
)
```
